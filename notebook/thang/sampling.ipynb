```python
import pandas as pd
import numpy as np

df = pd.read_csv('../../data/NASA Near-Earth Objects-Sampling.csv', index_col=0)


df.head()
```

```python
df['first_observation_date'] = pd.to_datetime(df['first_observation_date'])
df['last_observation_date'] = pd.to_datetime(df['last_observation_date'])
```

```python
df = df[~((df['last_observation_date'].dt.year < 2021) & (df['is_potentially_hazardous_asteroid'] == False))]
df.head()
```

```python
filename = '../../data/NASA Near-Earth Objects-Train.csv'
df.to_csv(filename,index=False)
```
